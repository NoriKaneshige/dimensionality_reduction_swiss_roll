# Dimensionality reduction, unrolling Swiss roll

```python
#Here we load in the swiss roll dataset
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets import make_swiss_roll
X, t= make_swiss_roll(n_samples=1000, noise=0.2, random_state=42)
# the data in t will be used for coloring

axes = [-11.5, 14, -2, 23, -12, 15]

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, cmap=plt.cm.hot)
ax.view_init(10, -70)
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
ax.set_zlabel("$x_3$", fontsize=18)
ax.set_xlim(axes[0:2])
ax.set_ylim(axes[2:4])
ax.set_zlim(axes[4:6])
plt.rcParams['axes.facecolor'] = 'olive'

plt.show()
```

```python
'''Apply Kernel PCA to the swiss roll dataset, namely:
(1)kernel="linear", 
(2)kernel="rbf", 
(3)kernel="sigmoid"
------
Plot these 3 kernel PCA results (using the 3 different kernels) and compare vs. Isomap. Use the first two principal components only. 
Use gamma=0.042 for rbf kernel and gamma = 0.0008 for sigmoid kernel'''
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.KernelPCA.html
# kernel“linear” | “poly” | “rbf” | “sigmoid” | “cosine” | “precomputed”
# Kernel. Default=”linear”.

from sklearn.decomposition import KernelPCA

# (1) linear KPCA
scikit_kpca = KernelPCA(n_components=2, kernel='linear', gamma=15)
X_skernpca = scikit_kpca.fit_transform(X)
# print(X_skernpca)

fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(111)

ax.scatter(X_skernpca[:, 0], X_skernpca[:, 1], c=t, cmap=plt.cm.hot)
# ax.view_init(10, -70)
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
# set_xlim(self, left=None, right=None, emit=True, auto=False, *, xmin=None, xmax=None)
ax.set_xlim(-15,15)
ax.set_ylim(-15,15)

plt.show()
```

```python
# (2) rbf KPCA
scikit_kpca = KernelPCA(n_components=2, kernel='rbf', gamma=0.042)
X_skernpca = scikit_kpca.fit_transform(X)
print(X_skernpca)

fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(111)

ax.scatter(X_skernpca[:, 0], X_skernpca[:, 1], c=t, cmap=plt.cm.hot)
# ax.view_init(10, -70)
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
# set_xlim(self, left=None, right=None, emit=True, auto=False, *, xmin=None, xmax=None)
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)

plt.show()
```

```python
# (3) sigmoid KPCA
scikit_kpca = KernelPCA(n_components=2, kernel='sigmoid', gamma=0.0008)
X_skernpca = scikit_kpca.fit_transform(X)
print(X_skernpca)

fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(111)

ax.scatter(X_skernpca[:, 0], X_skernpca[:, 1], c=t, cmap=plt.cm.hot)
# ax.view_init(10, -70)
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
# set_xlim(self, left=None, right=None, emit=True, auto=False, *, xmin=None, xmax=None)
ax.set_xlim(-0.5,0.5)
ax.set_ylim(-0.5,0.5)

plt.show()
```

What happens if you increase gamma to 0.1 for RBF and sigmoid kernels?

```python
#replot kernel PCA results using rbf kernel and sigmoid kernel but with gamma = 0.1.

# rbf KPCA replot
scikit_kpca = KernelPCA(n_components=2, kernel='rbf', gamma=0.1)
X_skernpca = scikit_kpca.fit_transform(X)
print(X_skernpca)

fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(111)

ax.scatter(X_skernpca[:, 0], X_skernpca[:, 1], c=t, cmap=plt.cm.hot)
# ax.view_init(10, -70)
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
# set_xlim(self, left=None, right=None, emit=True, auto=False, *, xmin=None, xmax=None)
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)

plt.show()

# sigmoid KPCA replot
scikit_kpca = KernelPCA(n_components=2, kernel='rbf', gamma=0.1)
X_skernpca = scikit_kpca.fit_transform(X)
print(X_skernpca)

fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(111)

ax.scatter(X_skernpca[:, 0], X_skernpca[:, 1], c=t, cmap=plt.cm.hot)
# ax.view_init(10, -70)
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
# set_xlim(self, left=None, right=None, emit=True, auto=False, *, xmin=None, xmax=None)
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
plt.rcParams['axes.facecolor'] = 'orange'

plt.show()
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.manifold.Isomap.html
# https://scikit-learn.org/stable/modules/manifold.html#isomap

#Plot isomap results here. Use n_components = 2
from sklearn.manifold import Isomap

# isomap
scikit_isomap = Isomap(n_components=2)
X_isomap = scikit_isomap.fit_transform(X)
print(X_isomap)

fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(111)

ax.scatter(X_isomap[:, 0], X_isomap[:, 1], c=t, cmap=plt.cm.hot)
# ax.view_init(10, -70)
ax.set_title("$x_1$", fontsize=18)
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
# set_xlim(self, left=None, right=None, emit=True, auto=False, *, xmin=None, xmax=None)
ax.set_xlim(-70,60)
ax.set_ylim(-50,50)
plt.rcParams['axes.facecolor'] = 'gray'

plt.show()
```
